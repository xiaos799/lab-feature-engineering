```python
import pandas as pd
```

```python
df = pd.read_csv("learningSet.csv")
df
```

```python
for col in df.columns:
    print(df[col].isna().sum())
```

```python
drop_list = []
drop_list.extend(['OSOURCE', 'ZIP'])
```

```python
drop_list
```

```python
drop_list.extend(col for col in df.columns if df[col].isna().sum()>= 95412 * 0.85)
drop_list
```

```python
df = df.drop(columns = drop_list)
df
```

```python
df['GENDER'].value_counts()
```

```python
df['GENDER'] = df['GENDER'].fillna('F')
df['GENDER'].value_counts()
```

```python
df['GENDER'] = df['GENDER'].apply(lambda x: 'other' if x not in ['F', 'M'] else x)
df['GENDER'].value_counts()
```

```python
df.info()
```

```python
numeric = df[[col for col in df.columns if df[col].dtype in ['float64', 'int64']]]
numeric
```

```python
for col in numeric.columns:
    print(numeric[col].isna().sum())
```

```python
toclean = df[['GEOCODE2', 'WEALTH1', 'ADI', 'DMA', 'MSA']]
toclean
```

```python
toclean.info()
```

```python
toclean.GEOCODE2 = toclean.GEOCODE2.fillna(toclean.GEOCODE2.value_counts().index[0])
toclean.GEOCODE2.info()
```

```python
toclean = toclean.drop('GEOCODE2', axis=1)
toclean.describe()
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3, weights='distance')
```

```python
toclean = imputer.fit_transform(toclean)
```

```python
toclean = pd.DataFrame({'WEALTH1': toclean[:, 0], 'ADI': toclean[:, 1], 'DMA': toclean[:, 2], 'MSA': toclean[:, 3]})
```
